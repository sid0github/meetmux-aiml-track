```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
import joblib

data = load_breast_cancer()
X,y = data.data, data.target

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    verbose=1
)

grid_search.fit(X, y)
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best cross-validation score: {grid_search.best_score_:.4f}")
print(f"Total Fits: {len(grid_search.cv_results_['params']) * 5}")

joblib.dump(grid_search.best_estimator_, 'best_model.pkl')
print("Model saved successfully")

loaded_model = joblib.load('best_model.pkl')
print(f"Loaded model prediction: {loaded_model.predict(X[:1])}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
X,y = data.data, data.target

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='precision',
    verbose=1
)

grid_search.fit(X, y)
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best cross-validation score: {grid_search.best_score_:.4f}")
print(f"Total Fits: {len(grid_search.cv_results_['params']) * 5}")
```

 "GridSearch tests every combination. If we have 5 parameters with 10 
options each, how many models will it train? Why use RandomizedSearchCV for 
huge datasets?" 
10*10*10*10*10=100000 combinations
100000*3=300000 models

RandomizedSearchCV tests 100 random combinations which gurantees speed which is optimal for huge datasets
